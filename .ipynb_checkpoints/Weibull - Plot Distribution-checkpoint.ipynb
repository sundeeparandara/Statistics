```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Weibull

https://en.wikipedia.org/wiki/Weibull_distribution

![eqn](weibull_eqn.png)

```python
def weibull_pdf(scale_parameter,shape_parameter,x):
    
    _lambda = scale_parameter 
    kappa = shape_parameter
    
    #pdf = kappa*(_lambda**kappa)*(x**(kappa-1))*np.exp((-1*(_lambda*x))**kappa)
    pdf = (kappa/_lambda)*((x/_lambda)**(kappa-1))*np.exp(-1*((x/_lambda)**kappa))
    
    return pdf
```

```python
def weibull_pdf_plot(scale_parameter,shape_parameter,x):
    
    _lambda = scale_parameter 
    kappa = shape_parameter
    
    pdf = (kappa/_lambda)*((x/_lambda)**(kappa-1))*np.exp(-1*((x/_lambda)**kappa))
    
    fig,ax = plt.subplots(figsize=(12,8))
    sns.lineplot(x=x,y=pdf,label=f"lambda={scale_parameter}, kappa={shape_parameter}")
```

```python
x = np.linspace(0,2,44)
scale_parameter = 1
shape_parameter_list = [0.5,1,2,3]
fig,ax = plt.subplots(figsize=(12,8))
for shape_parameter in shape_parameter_list:
    pdf = weibull_pdf(scale_parameter,shape_parameter,x)
    sns.lineplot(x=x,y=pdf,label=f"lambda={scale_parameter}, kappa={shape_parameter}")
```

```python
x = np.linspace(0,8,88)
scale_parameter_list = [1,2,3,4]
shape_parameter = 3.26
fig,ax = plt.subplots(figsize=(12,8))
for scale_parameter in scale_parameter_list:
    pdf = weibull_pdf(scale_parameter,shape_parameter,x)
    sns.lineplot(x=x,y=pdf,label=f"lambda={scale_parameter}, kappa={shape_parameter}")
```
